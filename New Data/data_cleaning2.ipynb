```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
pd.set_option('display.max_columns',None)
```

```python
balls = pd.read_csv('deliveries_updated_ipl_upto_2025.csv')
```

```python
balls.head()
```

```python
balls['date'] = pd.to_datetime(balls['date'])
```

```python
balls['isWide'] = balls['isWide'].fillna(0.0)
balls['isNoBall'] = balls['isNoBall'].fillna(0.0)
balls['player_dismissed'] = balls['player_dismissed'].fillna(0.0)
balls.head()
```

```python
balls['Byes'] = balls['Byes'].fillna(0)
balls['LegByes'] = balls['LegByes'].fillna(0)
balls['Penalty'] = balls['Penalty'].fillna(0)

balls['batsman_runs'] = balls['batsman_runs'] + balls['Byes'] + balls['LegByes']
```

```python
balls.drop(
    ['Byes','LegByes','over_ball','dismissal_kind','extras'],
    axis=1,
    inplace=True,
    errors='ignore'
)
```

```python
balls.head()
```

```python
balls['isWide'] = balls['isWide'] + balls['Penalty']
balls.drop(columns=['Penalty'], inplace=True)
balls.columns
```

```python
balls.head()
```

```python
balls['inning'].unique()
```

```python
balls = balls[balls['inning'].isin([1, 2])].copy()
balls['inning'] = balls['inning'].map({1: 0, 2: 1})
```

```python
print(balls.shape)
print(" -------------------- ")
print(balls.isnull().sum())
print(" -------------------- ")
print(balls.info())
```

```python
balls.to_csv('cleaned_deliveries_till_2025.csv')
```
