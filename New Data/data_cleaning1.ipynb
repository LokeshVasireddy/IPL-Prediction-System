```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
pd.set_option('display.max_columns',None)
```

```python
matches = pd.read_csv('matches_updated_ipl_upto_2025.csv')
```

```python
matches.head()
```

```python
matches.columns
```

```python
matches.drop(
    ['event','umpire2','toss_winner','neutralvenue','umpire1','reserve_umpire','match_referee','tv_umpire'],
    axis=1,
    inplace=True,
    errors='ignore'
)
matches.head()
```

```python
matches['season'].unique()
```

```python
matches['season'] = matches['season'].apply(
    lambda x: int(x.split('/')[0]) if x == '2020/21'
    else int('20' + x.split('/')[1]) if '/' in x
    else int(x)
)
matches['season'].unique()
```

```python
matches['winner_runs'] = matches['winner_runs'].fillna(0.0)
matches['winner_wickets'] = matches['winner_wickets'].fillna(0.0)
```

```python
matches['winner'] = matches['winner'].fillna(matches['eliminator'])
matches['winner'].unique()
```

```python
matches[matches['winner'].isna()][['matchId','winner','eliminator']]
```

```python
matches = matches[matches['winner'].notna()]
matches['winner'].isna().sum()
```

```python
matches.head()
```

```python
matches.drop(
    ['eliminator','date1','date2','toss_decision','method','gender','balls_per_over','outcome','city','match_number'],
    axis=1,
    inplace=True,
    errors='ignore'
)
```

```python
matches.head()
```

```python
print(matches.shape)
print(" -------------------- ")
print(matches.isnull().sum())
print(" -------------------- ")
print(matches.info())
```

```python
matches['date'] = pd.to_datetime(matches['date'])
```
